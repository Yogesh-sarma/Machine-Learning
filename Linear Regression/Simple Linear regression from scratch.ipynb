```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```
class simpleRegression:
  def __init__(self,learningRate=0.01,steps=1100): #learning rate is alpha #steps is no.of iterations
    self.learningRate=learningRate
    self.steps=steps

# gradient descent is used to optimise the equation y=b0 + b1(x)
  def fit(self,x,y):
    xtrain=np.c_[np.ones(x.shape[0]),x]
    self.theta=np.random.rand(xtrain.shape[1]) #theta is array with random b0 and b1 values
   
    for i in range(self.steps):
      self.theta=self.theta - (self.learningRate * self.calcDerivate(xtrain,y)) #repeat until convergence

  def calcDerivate(self,x,y): #see readme for this
    return (1/x.shape[0]) * (np.dot(x.T, (np.dot(x,self.theta) -y.reshape(-1))))

  def predict(self,x):
    #predicting new values using the calcualted b0 and b1 values in linear regression model
    x_pred=np.c_[np.ones(x.shape[0]),x]

    return np.dot(x_pred,self.theta)
```

```
#upload dataset
filename=pd.read_csv('/content/drive/My Drive/ML:a-z/Linear regression/Copy of Salary_Data.csv') 
data=filename.values

x=data[:,0:1] #independent variable
y=data[:,1:]  #dependent variable or target variable
#splitting data
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.2,random_state=0)
print(x_train.shape)
print(y_train.shape)
```

```
model= simpleRegression() #create instance
model.fit(x_train,y_train) #fitting the model
```

```
#plotting graph
plt.scatter(x_test,y_test,color='red')
plt.xlabel('Experience')
plt.ylabel('Salary')
plt.title('Salary vs Experience')
plt.plot(x_test,model.predict(x_test),color='black')
plt.show()
plt.xlabel('Exp')
plt.ylabel('sal')
plt.title('Training data')
plt.scatter(x_train,y_train,color='purple')
plt.plot(x_train,model.predict(x_train),color='orange')
plt.show()
```

# Polynomial Regression

## Importing the libraries

```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plot
```

## Importing the dataset

```
data=pd.read_csv('/content/drive/My Drive/ML:a-z/Polynomial Regression/Position_Salaries.csv')
x=data.iloc[:,1:-1].values #only levels is enough, no need of role name
y=data.iloc[:,-1].values
```

```
print(x)
```

## Training the Linear Regression model on the whole dataset

```
from sklearn.linear_model import LinearRegression
lin_reg=LinearRegression()
lin_reg.fit(x,y)
```

## Training the Polynomial Regression model on the whole dataset

```
from sklearn.preprocessing import PolynomialFeatures
poly_reg=PolynomialFeatures(degree=4)
x_poly=poly_reg.fit_transform(x)
polynomial_reg=LinearRegression()
polynomial_reg.fit(x_poly,y)
```

## Visualising the Linear Regression results

```
plot.xlabel('level')
plot.ylabel('Salary')
plot.title('Salary vs Level Linear Reg Model')
plot.scatter(x,y,color='red')
plot.plot(x,lin_reg.predict(x),color='blue')
plot.show()
```

## Visualising the Polynomial Regression results

```
plot.scatter(x,y,color='red')
plot.plot(x,polynomial_reg.predict(x_poly),color='green')
plot.xlabel('Level')
plot.ylabel('Salary')
plot.title('Salary vs Level Polynomial Reg Model')
plot.show()
```

## Visualising the Polynomial Regression results (for higher resolution and smoother curve)

```
X_grid = np.arange(min(x), max(x), 0.1)
X_grid = X_grid.reshape((len(X_grid), 1))
plot.scatter(x, y, color = 'red')
plot.plot(X_grid, polynomial_reg.predict(poly_reg.fit_transform(X_grid)), color = 'blue')
plot.title('Truth or Bluff (Polynomial Regression)')
plot.xlabel('Position level')
plot.ylabel('Salary')
plot.show()
```

## Predicting a new result with Linear Regression

```
plot.xlabel('level')
plot.ylabel('Salary')
plot.title('Salary vs Level Linear Reg Model')
plot.scatter(x,y,color='red')
plot.plot(x,lin_reg.predict(x),color='blue')
plot.scatter([[9.3]],lin_reg.predict([[9.3]]),color='orange')
plot.show()
```

## Predicting a new result with Polynomial Regression

```
X_grid = np.arange(min(x), max(x), 0.1)
X_grid = X_grid.reshape((len(X_grid), 1))
plot.scatter(x, y, color = 'red')
plot.plot(X_grid, polynomial_reg.predict(poly_reg.fit_transform(X_grid)), color = 'blue')
plot.title('Truth or Bluff (Polynomial Regression)')
plot.xlabel('Position level')
plot.ylabel('Salary')
plot.scatter([[9.3]],polynomial_reg.predict(poly_reg.fit_transform([[9.3]])),color='orange')
plot.show()
```
